### Image read and show using CV

```python
import cv2
```

```python
#path must be correctly declared,
#otherwise there no error will be appeared

img = cv2.imread('DATA/00-puppy.jpg')
```

```python
#img
plt.imshow(img)
```

```python
type(img)
```

```python
img.shape
```

The image has been correctly loaded by openCV as a numpy array, but the color of each pixel has been sorted as BGR. Matplotlib's plot expects an RGB image so, for a correct display of the image, it is necessary to swap those channels. This operation can be done either by using openCV conversion functions cv2.cvtColor() or by working directly with the numpy array.

**MATPLOTLIB -->  RGB RED GREEN BLUE**

**OPENCV -->  BGR BLUE GREEN RED**

#### Convert BGR to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

#### Convert in gray

```python
img_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray, cmap='gray')
```

## Resize Images

```python
img_rgb.shape
# width, height, color channels
```

```python
resized_img = cv2.resize(img_rgb, (1300, 275))
resized_img.shape
```

```python
plt.imshow(resized_img)
```

### By Ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_resized_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
new_resized_img.shape
```

```python
plt.imshow(new_resized_img)
```

```python
%matplotlib qt
plt.imshow(new_resized_img)
```

### Flipping Images

```python
%matplotlib inline
```

```python
# Along central x axis
new_img = cv2.flip(new_resized_img, 0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_resized_img, -1)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_resized_img, 1)
plt.imshow(new_img)
```

### Saving Image Files

```python
type(new_img)
```

```python
cv2.imwrite('DATA/my_new_picture.jpg',new_img)
```
